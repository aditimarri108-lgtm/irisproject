```python
import pandas as pd
```

```python
df=pd.read_csv('Iris.csv')
```

```python
df.head(2)
```

```python
df.info()
```

```python
df['Species'].unique()   # for unique categories
```

```python
# separating input and output features

X=df.drop(labels=['Id','Species'],axis=1)
Y=df[['Species']]      output features
```

```python
X
```

```python
# feature scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler().set_output(transform='pandas')
X_pre= scaler.fit_transform(X) # fitting the data X for scaling
```

```python
import pickle  # pickle is used as a bridge to connect frontend and backend
with open('scaler.pkl','wb') as file1:
    pickle.dump(scaler,file1)
```

```python
X_pre
```

```python
# splitting the input data into training and testing
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X_pre,Y,test_size=0.33,randomstate=20)
```

```python
xtest.shape()
```

```python
ytest.value_counts()
```

```python
# model bilding
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
# evaluation
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
with open('model.pkl','wb') as file2:
    pickle.dump(model,file2)
```
